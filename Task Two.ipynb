#### Predefined Financial Queries for Chatbot
1. What is the total revenue?
2. What is the Net Income?
3. What is the sum of total assets?
4. What is the sum of total liabilities?
5. What is cash flow from operating activities?
6. What is the revenue growth(%) ?
7. What is the net income growth(%) ?
8. What is the assets growth(%) ?
9. What is the liabilities growth(%) ?
10. What is the cash flow from operations growth(%) ?
11. What is the year by year average revenue growth rate(%)?
12. What is the year by year average net income growth rate(%)?
13. What is the year by year average assets growth rate(%)?
14. What is the year by year average liabilities growth rate(%)?
15. What is the year by year average cash flow from operations growth rate(%)?

```python
def get_company_data(df, company, year):
    """
    Helper function to extract the financial data for the specific company and year.
    """
    company_data = df[(df['Company'] == company) & (df['Year'] == year)]
    return company_data

def simple_chatbot():
    while True:
        user_input = input("\nEnter Hi to start the chatbot session; type 'exit' to quit): ")
        if user_input.lower() == "exit":
            break
        elif user_input.lower() == "hi":
            print("\nHello! Welcome to AI Driven Financial Chatbot!!!")
            print("\nI can help you with your financial queries")
            print("Please select the company name from below: -")
            print("\n1.Microsoft \n2.Tesla \n3.Apple")
            company_name = input("Please select the company name from below (Microsoft, Tesla, Apple): ")
            year = int(input("Please enter the year (2021, 2022, or 2023): "))
            company_data = get_company_data(df, company_name, year)
    
            if company_data.empty:
                print(f"No data found for {company_name} in {year}. Please try again.")
                return

            user_query = input("Please enter your query (e.g., 'What is the total revenue?', 'What is the assets growth?'): ").lower()

            if "total revenue" in user_query:
                print(f"The total revenue for {company_name} in {year} is ${company_data['Total Revenue'].values[0]:,.2f}.")
    
            elif "net income" in user_query:
                print(f"The net income for {company_name} in {year} is ${company_data['Net Income'].values[0]:,.2f}.")
    
            elif "total assets" in user_query:
                print(f"The total assets for {company_name} in {year} is ${company_data['Total Assets'].values[0]:,.2f}.")
    
            elif "total liabilities" in user_query:
                print(f"The total liabilities for {company_name} in {year} is ${company_data['Total Liabilities'].values[0]:,.2f}.")
    
            elif "cash flow from operating activities" in user_query:
                print(f"The cash flow from operating activities for {company_name} in {year} is ${company_data['Cash Flow from Operating Activities'].values[0]:,.2f}.")
    
            elif "revenue growth" in user_query:
                print(f"The revenue growth (%) for {company_name} in {year} is {company_data['Revenue Growth (%)'].values[0]:.2f}%.")
    
            elif "net income growth" in user_query:
                print(f"The net income growth (%) for {company_name} in {year} is {company_data['Net Income Growth (%)'].values[0]:.2f}%.")
    
            elif "assets growth" in user_query:
                print(f"The assets growth (%) for {company_name} in {year} is {company_data['Assets Growth (%)'].values[0]:.2f}%.")
    
            elif "liabilities growth" in user_query:
                print(f"The liabilities growth (%) for {company_name} in {year} is {company_data['Liabilities Growth (%)'].values[0]:.2f}%.")
    
            elif "cash flow growth" in user_query:
                print(f"The cash flow from operations growth (%) for {company_name} in {year} is {company_data['Cash Flow from Operations Growth(%)'].values[0]:.2f}%.")
    
            else:
                print("Sorry, I can only provide information on predefined queries.")

simple_chatbot()
```
